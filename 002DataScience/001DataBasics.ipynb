## CREATING DATA FRAMES
[Back To Table of Content](#table-of-content)

```python
import pandas as pd

# data
data = {
    "name" : ["Adam", "Betty", "Celiwe", "Delani"],
    "age" : [18, 19, 24, 30]
}

# Loads data into dataframe
df = pd.DataFrame(data)

# View the data
print(df)
```

```python
# print by row index
print(df.loc[0])
```
